```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
my_img1=cv2.imread("1.png",0)
my_kernel = np.ones((5,5),dtype = np.uint8)
my_kernel
```

```python
my_erosion_img = cv2.erode(my_img1 , my_kernel, iterations = 2)
fig = plt.figure(figsize = (10,7))
rows=1
columns=2
fig.add_subplot(rows,columns,1)
plt.title("original_resim")
plt.imshow(my_img1,cmap="gray")
plt.axis("off")



fig.add_subplot(rows,columns,2)
plt.title("erosion_resim")
plt.imshow(my_erosion_img, cmap="gray")
plt.axis("off")
```

```python
my_dilation_img = cv2.dilate(my_img1 , my_kernel, iterations = 2)
fig = plt.figure(figsize = (10,7))
rows=1
columns=3
fig.add_subplot(rows,columns,1)
plt.title("original_resim")
plt.imshow(my_img1,cmap="gray")
plt.axis("off")



fig.add_subplot(rows,columns,2)
plt.title("erosion_resim")
plt.imshow(my_erosion_img, cmap="gray")
plt.axis("off")


fig.add_subplot(rows,columns,3)
plt.title("dilation_resim")
plt.imshow(my_dilation_img, cmap="gray")
plt.axis("off")
```

```python
my_img2 = cv2.imread("2.png",0)
white_noise = np.random.randint(0,2,size=my_img2.shape[:2])
white_noise*=255

fig = plt.figure(figsize = (10,7))
rows=1
columns=2
fig.add_subplot(rows,columns,1)
plt.title("original_resim")
plt.imshow(my_img2,cmap="gray")
plt.axis("off")



fig.add_subplot(rows,columns,2)
plt.title("white_noise_resim")
plt.imshow(white_noise, cmap="gray")
plt.axis("off")
```

```python
my_noised_image = my_img2+white_noise
plt.imshow(my_noised_image,cmap="gray")
```

```python
my_opening_img = cv2.morphologyEx(my_noised_image.astype(np.float32), cv2.MORPH_OPEN,my_kernel)

fig = plt.figure(figsize = (10,7))
rows=1
columns=4
fig.add_subplot(rows,columns,1)
plt.title("original_resim")
plt.imshow(my_img2,cmap="gray")
plt.axis("off")



fig.add_subplot(rows,columns,2)
plt.title("white_noise_resim")
plt.imshow(white_noise, cmap="gray")
plt.axis("off")

fig.add_subplot(rows,columns,3)
plt.title("noised_resim")
plt.imshow(my_noised_image, cmap="gray")

plt.axis("off")

fig.add_subplot(rows,columns,4)
plt.title("opening_resim")
plt.imshow(my_opening_img, cmap="gray")
plt.axis("off")
```

```python
black_noise = np.random.randint(0,2,size=my_img2.shape[:2])
black_noise*=-255
```

```python
fig = plt.figure(figsize = (10,7))
rows=1
columns=5
fig.add_subplot(rows,columns,1)
plt.title("original_resim")
plt.imshow(my_img2,cmap="gray")
plt.axis("off")



fig.add_subplot(rows,columns,2)
plt.title("white_noise_resim")
plt.imshow(white_noise, cmap="gray")
plt.axis("off")

fig.add_subplot(rows,columns,3)
plt.title("noised_resim")
plt.imshow(my_noised_image, cmap="gray")

plt.axis("off")

fig.add_subplot(rows,columns,4)
plt.title("opening_resim")
plt.imshow(my_opening_img, cmap="gray")
plt.axis("off")

fig.add_subplot(rows,columns,5)
plt.title("black_noise_resim")
plt.imshow(black_noise, cmap="gray")
plt.axis("off")
```
