```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
my_img_rgb = cv2.imread("./aslan.jpg",1)
my_img_gray = cv2.cvtColor(my_img_rgb,cv2.COLOR_BGR2GRAY)
```

```python
my_img_gray
```

```python
"""
ADAPTIVE_THRESH_MEAN_C = adaptif eşikleme yöntemi türü
THRESH_BINARY = Adaptif eşikleme işleminde kullanılacak (baz alınacak)
normal eşikleme türü
255 = en büyük piksel değeri (eşiklemeden sonra verilecek en büyük piksel
değeri
11 = blok büyüklüğü (11 satır 11 sütunluk bir matris ile görüntü
parçalara ayrılacak ve her parça içerisinde lokal eşikleme yapılacak
8 = eşiklemeden soran piksel değerlerinin çıkarılacağı sayı
"""
my_adaptive_binary_img = cv2.adaptiveThreshold(my_img_gray, 255, 
            cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 11, 8)
```

```python
fig = plt.figure(figsize=(10, 7))
rows = 1
columns = 2

fig.add_subplot(rows, columns, 1)
plt.imshow(my_img_gray, cmap="gray")
plt.axis('off')
plt.title("Original Image")

fig.add_subplot(rows, columns, 2)
plt.imshow(my_adaptive_binary_img, cmap="gray")
plt.axis('off')
plt.title("ADAPTIVE BINARY THRESHOLDING Image")
```

```python
my_adaptive_binary_inv_img = cv2.adaptiveThreshold(my_img_gray, 255, 
            cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY_INV, 11, 8)
```

```python
fig = plt.figure(figsize=(10, 7))
rows = 1
columns = 3

fig.add_subplot(rows, columns, 1)
plt.imshow(my_img_gray, cmap="gray")
plt.axis('off')
plt.title("Original Image")

fig.add_subplot(rows, columns, 2)
plt.imshow(my_adaptive_binary_img, cmap="gray")
plt.axis('off')
plt.title("ADAPTIVE BINARY THRESHOLDING Image")

fig.add_subplot(rows, columns, 3)
plt.imshow(my_adaptive_binary_inv_img, cmap="gray")
plt.axis('off')
plt.title("ADAPTIVE BINARY INV THRESHOLDING Image")
```

```python
my_adaptive_binary_gauss_img = cv2.adaptiveThreshold(my_img_gray, 255, 
            cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 11, 8)
```

```python
fig = plt.figure(figsize=(18, 7))
rows = 1
columns = 3

fig.add_subplot(rows, columns, 1)
plt.imshow(my_img_gray, cmap="gray")
plt.axis('off')
plt.title("Original Image")

fig.add_subplot(rows, columns, 2)
plt.imshow(my_adaptive_binary_img, cmap="gray")
plt.axis('off')
plt.title("ADAPTIVE MEAN BINARY THRESHOLDING Image")

fig.add_subplot(rows, columns, 3)
plt.imshow(my_adaptive_binary_gauss_img, cmap="gray")
plt.axis('off')
plt.title("ADAPTIVE BINARY GAUSS THRESHOLDING Image")
```
