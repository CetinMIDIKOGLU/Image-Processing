```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
my_img = cv2.imread("./aslan.jpg",1)
my_img_2 = cv2.cvtColor(my_img,cv2.COLOR_BGR2GRAY)
```

```python
#EŞİKLEME (THRESHOLDING)
print(my_img.shape)
print(my_img_2.shape)
```

```python
# thresholding işlemi, verilen eşik değerine göre o değerin altında kalan
# piksellerin sıfıra (siyah renge) ya da ters işlemde 255'e (beyaza)
# dönüştürmeye yarar.

_, my_binary_thresh_img = cv2.threshold(my_img_2, thresh=125, maxval=255, 
                                        type=cv2.THRESH_BINARY)
```

```python
print(_)
```

```python
plt.imshow(my_binary_thresh_img, cmap="gray")
plt.axis("off")
```

```python
fig = plt.figure(figsize=(10, 7))
rows = 1
columns = 2

fig.add_subplot(rows, columns, 1)
plt.imshow(my_img_2, cmap="gray")
plt.axis('off')

fig.add_subplot(rows, columns, 2)
plt.imshow(my_binary_thresh_img, cmap="gray")
plt.axis('off')
```

```python
_, my_binary_thresh_inv_img = cv2.threshold(my_img_2, thresh=125, maxval=255, 
                                        type=cv2.THRESH_BINARY_INV)
```

```python
fig = plt.figure(figsize=(12, 7))
rows = 1
columns = 3

fig.add_subplot(rows, columns, 1)
plt.imshow(my_img_2, cmap="gray")
plt.axis('off')
plt.title("Original Image")

fig.add_subplot(rows, columns, 2)
plt.imshow(my_binary_thresh_img, cmap="gray")
plt.axis('off')
plt.title("BINARY THRESHOLDING Image")

fig.add_subplot(rows, columns, 3)
plt.imshow(my_binary_thresh_inv_img, cmap="gray")
plt.axis('off')
plt.title("BINARY INVERSE THRESHOLDING Image")
```

```python
# THRESH_TRUNC: Eşik değerinin üstündeki pikseller eşik değerine kesilir (eşitlenir), 
# altındaki pikseller aynı kalır.

_, my_binary_thresh_trunc_img = cv2.threshold(my_img_2, thresh=125, maxval=255, 
                                        type=cv2.THRESH_TRUNC)
```

```python
fig = plt.figure(figsize=(8, 5))
rows = 2
columns = 2

fig.add_subplot(rows, columns, 1)
plt.imshow(my_img_2, cmap="gray")
plt.axis('off')
plt.title("Original Image")

fig.add_subplot(rows, columns, 2)
plt.imshow(my_binary_thresh_img, cmap="gray")
plt.axis('off')
plt.title("BINARY THRESHOLDING Image")

fig.add_subplot(rows, columns, 3)
plt.imshow(my_binary_thresh_inv_img, cmap="gray")
plt.axis('off')
plt.title("BINARY INVERSE THRESHOLDING Image")

fig.add_subplot(rows, columns, 4)
plt.imshow(my_binary_thresh_trunc_img, cmap="gray")
plt.axis('off')
plt.title("TRUNC THRESHOLDING Image")
```

```python
# TOZERO : Eşik değerinin altındaki pikseller 0 olur, üstündeki pikseller aynı kalır.

_, my_binary_thresh_tozero_img = cv2.threshold(my_img_2, thresh=125, maxval=255, 
                                        type=cv2.THRESH_TOZERO)
```

```python
# TOZERO : Eşik değerinin altındaki pikseller 255 olur, üstündeki pikseller aynı kalır.
_, my_binary_thresh_tozero_inv_img = cv2.threshold(my_img_2, thresh=125, maxval=255, 
                                        type=cv2.THRESH_TOZERO_INV)
```

```python
fig = plt.figure(figsize=(8, 5))
rows = 3
columns = 2

fig.add_subplot(rows, columns, 1)
plt.imshow(my_img_2, cmap="gray")
plt.axis('off')
plt.title("Original Image")

fig.add_subplot(rows, columns, 2)
plt.imshow(my_binary_thresh_img, cmap="gray")
plt.axis('off')
plt.title("BINARY THRESHOLDING Image")

fig.add_subplot(rows, columns, 3)
plt.imshow(my_binary_thresh_inv_img, cmap="gray")
plt.axis('off')
plt.title("BINARY INVERSE THRESHOLDING Image")

fig.add_subplot(rows, columns, 4)
plt.imshow(my_binary_thresh_trunc_img, cmap="gray")
plt.axis('off')
plt.title("TRUNC THRESHOLDING Image")

fig.add_subplot(rows, columns, 5)
plt.imshow(my_binary_thresh_tozero_img, cmap="gray")
plt.axis('off')
plt.title("TOZERO THRESHOLDING Image")

fig.add_subplot(rows, columns, 6)
plt.imshow(my_binary_thresh_tozero_inv_img, cmap="gray")
plt.axis('off')
plt.title("TOZERO INV THRESHOLDING Image")
```
