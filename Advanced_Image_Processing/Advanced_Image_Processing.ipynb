```python
import cv2
from matplotlib import pyplot as plt 
import numpy as np
```

```python
img_original = cv2.imread("1.jpeg")
```

```python
plt.imshow(img_original)
```

```python
img_rgb = cv2.cvtColor(img_original, cv2.COLOR_BGR2RGB)
img_gray = cv2.cvtColor(img_original, cv2.COLOR_BGR2GRAY)
```

```python
fig = plt.figure(figsize = (10,7))
rows=1
columns=3
fig.add_subplot(rows,columns,1)
plt.title("original_resim")
plt.imshow(img_original)
plt.axis("off")

fig.add_subplot(rows,columns,2)
plt.title("rgb_resim")
plt.imshow(img_rgb)
plt.axis("off")

fig.add_subplot(rows,columns,3)
plt.title("gray_resim")
plt.imshow(img_gray,cmap="gray")
plt.axis("off")
```

```python
img_hsv_green = cv2.cvtColor(img_original, cv2.COLOR_BGR2HSV)

lower_green = np.array([20,100,90])
upper_green = np.array([130,210,130])

mask_green = cv2.inRange(img_hsv_green, lower_green, upper_green)
```

```python
fig = plt.figure(figsize = (10,7))
rows=1
columns=2
fig.add_subplot(rows,columns,1)
plt.title("img_rgb_new_1")
plt.imshow(img_rgb_new_1)
plt.axis("off")

fig.add_subplot(rows,columns,2)
plt.title("mask_green")
plt.imshow(mask_green,cmap="gray")
plt.axis("off")

```

```python
yellow =np.full_like(img_rgb_new_1, [255,255,0])

img_rgb_new_1[mask_green !=0] = yellow[mask_green !=0]
```

```python
height, width = 100,100
img_red = np.zeros((height,width,3),dtype=np.uint8)
img_red[:,:,0] = 255

img_green = np.zeros((height,width,3),dtype=np.uint8)
img_green[:,:,1]=255

img_blue = np.zeros((height,width,3),dtype=np.uint8)
img_blue[:,:,2]=255

img_yellow= np.zeros((height,width,3),dtype=np.uint8)
img_yellow[:,:,:2]=255


img_black =np.zeros((height,width,3),dtype=np.uint8)

img_white =np.ones((height,width,3),dtype=np.uint8)*255

fig = plt.figure(figsize = (10,7), facecolor='purple')
rows=1
columns=6
fig.add_subplot(rows,columns,1)
plt.title("red")
plt.imshow(img_red)
plt.axis("off")

fig.add_subplot(rows,columns,2)
plt.title("green")
plt.imshow(img_green)
plt.axis("off")

fig.add_subplot(rows,columns,3)
plt.title("blue")
plt.imshow(img_blue)
plt.axis("off")

fig.add_subplot(rows,columns,4)
plt.title("yellow")
plt.imshow(img_yellow)
plt.axis("off")

fig.add_subplot(rows,columns,5)
plt.title("black")
plt.imshow(img_black)
plt.axis("off")

fig.add_subplot(rows,columns,6)
plt.title("white")
plt.imshow(img_white)
plt.axis("off")
```
